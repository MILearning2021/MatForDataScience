```python
#уровень 0
#Посчитайте значение первой и второй производной функции:
#f(x) = x^5 + 4sin(2x) + cos(3x+3)
#В точке x = 1.

import scipy as scp
import numpy as np

def f(x) :
    return x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)

x = 1.0
print("Первая производная: " + str(scp.misc.derivative(f, x, n = 1)))
print("Вторая производная: " + str(scp.misc.derivative(f, x, n = 2)))
```

```python
#Уровень 1
#Постройте график первой и второй производной функции:
#f(x) = sin^5(2x+1)
#На отрезке [-5; 5].

import matplotlib.pyplot as plt

def f(x):
    return (np.sin(2*x + 1))**5

axiesX = np.linspace(-5, 5, num = 10)
#точками
plt.subplot(121)
for i in range(0, len(axiesX), 1):
    f1 = scp.misc.derivative(f, axiesX[i], n = 1)
    f2 = scp.misc.derivative(f, axiesX[i], n = 2)    
    plt.plot(axiesX[i],f1,'o')
    plt.plot(axiesX[i],f2,'b*')

#линиями
plt.subplot(122)
f1 = scp.misc.derivative(f, axiesX, n = 1)
f2 = scp.misc.derivative(f, axiesX, n = 2)
plt.plot(axiesX,f1,'b--')
plt.plot(axiesX,f2,'r-.')
```

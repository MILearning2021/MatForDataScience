```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

from ipywidgets import IntProgress
from IPython.display import display
import time
```

```python
#       , 
#   n  1000   n      .

N = [2, 5, 10, 30, 50]

S = [0.0] * len(N)
for i in range(len(S)): S[i] = [0.0] * 1000

progress = IntProgress(min=0, max=1000, value=0, description='Sn, n=', bar_style='info')    
display(progress)
    
rv = sts.norm(loc=0, scale=3)
for n in range(len(N)):
    progress.value = 0
    progress.description = 'Sn, n=' + str(N[n])
    for i in range(0,1000):
        sample = rv.rvs(size = N[n])
        S[n][i] = sum(sample) / N[n]
        progress.value =  i + 1

for n in range(len(N)):
    plt.figure(figsize=(10, 5))
    plt.grid(ls='-')
    plt.xlim(-7,7)
    plt.ylim(None,1.2)
    plt.hist(S[n], density=True, bins=30, label=N[n])
    plt.legend()
    plt.show()
    
plt.figure(figsize=(10, 5))
plt.grid(ls='-')
plt.xlim(-7,7)
plt.ylim(None,1.2)
for n in range(len(N)):
    plt.hist(S[n], density=True, bins=30, label=N[n])

plt.legend()
plt.show()
```

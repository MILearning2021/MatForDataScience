```python
#Уровень 0:
#Задание 1:
#Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с 
#разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps

color = ['blue','green','red','yellow','magenta']

#нормальное
axiesX = np.linspace(-5, 5, num = 100)
plt.figure(figsize=(10, 5))
plt.grid(ls='-')
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)

for i in range(0, 5, 1):
    loc = round(np.random.uniform(-3,3), 1)
    scale = round(np.random.uniform(0.1,1),1)
    label = color[i] + ': мю=' + str(loc) + ', сигма=' + str(scale)
    print(label)
    plt.plot(axiesX, sps.norm.pdf(axiesX, loc = loc, scale = scale), '--', color = color[i], lw = 2)

plt.show()
```

```python
#экспоненциальное

axiesX = np.linspace(0, 3, 100)
plt.figure(figsize=(10, 5))
plt.grid(ls='-')
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)

loc = [0,0,0,0.75,1.5]

for i in range(0, 5, 1):
    scale = round(np.random.uniform(1,5),1)
    label = color[i] + ': параметр "scale"=' + str(scale)
    print(label)
    plt.plot(axiesX, sps.expon.pdf(axiesX, loc = loc[i], scale = scale), '--', color = color[i], lw = 2)
    
plt.show()

```

```python
#Стьюдента

axiesX = np.linspace(-5, 5, 1000)
plt.figure(figsize=(10, 5))
plt.grid(ls='-')
plt.xlabel('Значение', fontsize=18)
plt.ylabel('Плотность', fontsize=18)

df = 0.2

for i in range(0, 5, 1):
    loc = round(np.random.uniform(-3,3), 1)
    label = color[i] + ': мю=' + str(loc) + ', df=' + str(df+i)
    print(label)
    plt.plot(axiesX, sps.t.pdf(axiesX,  df + i/5, loc = loc, scale = 1), '--', color = color[i], lw = 2)
    
plt.show()
```

```python
#Уровень 1:
#Задание 2
#Непрерывная случайная величина задана плотностью распределения:
#f(xi)=C, xi in [0;5];   f(xi)=0, xi notin [0;5]
#Найдите C, математическое ожидание xi а также вероятность попадания в отрезок [3.5;7], т.е. P(3.5 < xi < 7).

from scipy.integrate import quad

def func(x,y): return y

#найти C, равномерное распределение
a = 0
b = 5
C = 1 / (b- a)
#найти МОЖ
E = (b + a) / 2
#найти вероятность попадания в отрезок
#P=P[3.5,5]+P[5,7]
P = quad(func, 3.5, 5.0, args=(C)) + quad(func, 3.5, 5.0, args=(0))

print("C = " + str(C))
print("E = " + str(E))
print("P = " + str(round(P[0],2)))
```

```python
#Задание 3
#Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. 
#Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. 
#А что будет в случае 28, 28 и 28 секунд для каждого цвета?

import math

t_l = [54., 4., 28.]

#случай 1
t_sum = sum(t_l)
I = -1 * ((t_l[0]/t_sum) * math.log2(t_l[0]/t_sum) + (t_l[1]/t_sum) * math.log2(t_l[1]/t_sum) + 
          (t_l[2]/t_sum) * math.log2(t_l[2]/t_sum))
print("Случай №1 - содержится " + str(int(math.ceil(I))) + " бит информации.")

#случай 2(равновероятные события)
I = abs(math.log2(1 / len(t_l)))
print("Случай №2 - содержится " + str(int(math.ceil(I))) + " бит информации.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Создайте случайную величину из любого выбранного вами распределения, 
#для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

N = [2, 5, 10, 30]

S = [0.0] * 4
for i in range(len(S)): S[i] = [0.0] * 1000

rv = sts.norm(loc=0, scale=3)
for n in range(len(N)):
    for i in range(0,1000):
        sample = rv.rvs(size = N[n])
        S[n][i] = sum(sample) / N[n]

for n in range(len(N)):
    plt.figure(figsize=(10, 5))
    plt.grid(ls='-')
    plt.xlim(-8,8)
    plt.ylim(None,0.8)
    plt.hist(S[n], density=True, bins=30, label=N[n])
    plt.legend()
    plt.show()
    
plt.figure(figsize=(10, 5))
plt.grid(ls='-')
for n in range(len(N)):
    plt.hist(S[n], density=True, bins=30, label=N[n])

plt.legend()
plt.show()
```
